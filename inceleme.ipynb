Dosyalarda duplicate veriye rastlamadım. Son olarak eksik değerlerin ne kadar olduğuna bakacağım.

```python
print(actors.isnull().sum())
print(countries.isnull().sum())
print(genres.isnull().sum())
print(languages.isnull().sum())
print(movies.isnull().sum())
print(movies2actors.isnull().sum())
print(prodcompanies.isnull().sum())
print(ratings.isnull().sum())
```

**Ülkelere göre çekilen film sayısı**

```python
plt.figure(figsize=(15,5))
plt.ylabel('Çekilen film sayısı')
countries["country"].value_counts().plot.bar(width= 0.5, color="#87CEEB")
```

**Aktörlerin oynadığı film sayısı**

```python
plt.figure(figsize=(15,5))
plt.ylabel('Aktörün oynadığı film sayısı')
actors["name"].value_counts().plot()
```

**Kadın ve Erkek Aktör Sayısı**

```python
movies2actors_played= (movies
                  .set_index('movieid')
                  .join(actors.set_index('actorid'),
                        how='inner')
                 )
movies2actors_played
```

```python
movies2actors_played.value_counts().plot.pie(autopct='%1.0f%%', figsize=(10, 10), textprops={'color':"w"})
```
